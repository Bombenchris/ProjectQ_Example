```python
from projectq import MainEngine
from projectq.ops import *
```

```python
def NOT(input):
    
    eng = MainEngine()
    qubit = eng.allocate_qubit()

    # consider input as a number or string input. 
    if input == 1 or input == '1':
        X | qubit
    
    X | qubit
    Measure | qubit
    
    eng.flush()
    output = int(qubit)
    
    print("Measured {}".format(int(qubit)))
    return output
```

```python
def XOR(input1, input2):
    eng = MainEngine()
    qureg = eng.allocate_qureg(2)
    
    if input1 == 1 or input1 == '1':
        X | qureg[0]
    if input2 == 1 or input2 == '1':
        X | qureg[1]

    CNOT | (qureg[0], qureg[1])
    Measure | qureg[1]
    
    eng.flush()
    output = int(qureg[1])
    print("Measured {}".format(int(qureg[1])))  
    
    return output
    
    
```

```python
def AND(input1, input2): 
    eng = MainEngine()
    qureg = eng.allocate_qureg(3)

    if input1 == 1 or input1 == '1':
        X | qureg[0]
    if input2 == 1 or input2 == '1':
        X | qureg[1]
    
    Toffoli | (qureg[0],qureg[1],qureg[2])
    Measure | qureg[2]
    
    eng.flush()
    output = int(qureg[2])
    print("Measured {}".format(int(qureg[2])))  
    
    return output
    
```

```python
def NAND(input1,input2):
    eng = MainEngine()
    qureg = eng.allocate_qureg(3)

    X | qureg[2]
    
    if input1 == 1 or input1 == '1':
        X | qureg[0]
    if input2 == 1 or input2 == '1':
        X | qureg[1]
    
    Toffoli | (qureg[0],qureg[1],qureg[2])
    Measure | qureg[2]
    
    eng.flush()
    output = int(qureg[2])
    print("Measured {}".format(int(qureg[2])))  
    
    return output   
```

```python
def OR(input1,input2):
    
    eng = MainEngine()
    qureg = eng.allocate_qureg(3)
    
    if input1 == 1 or input1 == '1' or input2 == 1 or input2 == '1':
        X | qureg[2]

    Measure | qureg[2]
    eng.flush()
    output = int(qureg[2])
    print("Measured {}".format(int(qureg[2])))  
    
    return output   
        
        
```

```python
print('\nResults for the NOT gate!')
for input in ['0','1']:
    print('   Input',input,'output',NOT(input))

print('\nResults for the XOR gate')
for input1 in ['0','1']:
    for input2 in ['0','1']:
        print('   Input',input1,input2,'output',XOR(input1,input2))
        
print('\nResults for the AND gate')
for input1 in ['0','1']:
    for input2 in ['0','1']:
        print('   Input',input1,input2,'output',AND(input1,input2))
        
print('\nResults for the NAND gate')
for input1 in ['0','1']:
    for input2 in ['0','1']:
        print('   Input',input1,input2,'output',NAND(input1,input2))
        
print('\nResults for the OR gate')
for input1 in ['0','1']:
    for input2 in ['0','1']:
        print('   Input',input1,input2,'output',OR(input1,input2))
```
